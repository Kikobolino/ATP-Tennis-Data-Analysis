## Initial Preparations and Data Import

```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
# Defining path for data export
path = r'C:\Users\chris\OneDrive\Desktop\Data Analytics CF\06_Advanced Analytics\02_Prepared Data'
```

```python
# Setting file location
files = glob.glob('C:/Users/chris/OneDrive/Documents/GitHub/tennis_atp/Matchdata/*.csv')
```

```python
# creating empty variable for list comprehension
frame = []
```

```python
# Adding file names in list comprehension
for file in files:
    frame.append(pd.read_csv(file))
```

```python
# Concatenating frames
df = pd.concat(frame, ignore_index = True)
```

```python
# Setting max display of columns to unlimited since the dataframe has 49 columns, to get an initial overview
pd.set_option("display.max_columns", None)
```

### Adjusting Data Types

```python
df['tourney_date'] = pd.to_datetime(df['tourney_date'], format='%Y%m%d')
```

```python
df['winner_id'] = df['winner_id'].astype(str)
df['loser_id'] = df['loser_id'].astype(str)
```

```python
df.loc[df['winner_age'].isna()]
# Only two winning players/matches without age no adjustment made
```

```python
df.loc[df['minutes'] >= 400]
# Match duration is wrong for rows 43624 and 47590 as row number 27379 is the longest match in tennis history
```

```python
# Adjusting wrong minutes for matches above
df.loc[(df['tourney_id'] == '2016-M001') & (df['match_num'] == 294), 'minutes'] = 90
df.loc[(df['tourney_id'] == '2017-0308') & (df['match_num'] == 297), 'minutes'] = 125
```

```python
# Checking result 
df.loc[(df['tourney_id'] == '2016-M001') & (df['match_num'] == 294)]
```

```python
df.loc[df['match_num'] > 1000]
# Investigating match number for Roland Garros 2019 & 2020 -> numbering indicates Round on 2nd and match number of round for 3rd and 4th digit
```

```python
# Checking for mixed type columns
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
# Adjusting values LL, Q and WC to be winner_entry values instead of winner_seed
df.loc[df['winner_seed'] == 'LL', 'winner_entry'] = 'LL'
df.loc[df['winner_seed'] == 'LL', 'winner_seed'] = float('NaN')
df.loc[df['winner_seed'] == 'Q', 'winner_entry'] = 'Q'
df.loc[df['winner_seed'] == 'Q', 'winner_seed'] = float('NaN')
df.loc[df['winner_seed'] == 'WC', 'winner_entry'] = 'WC'
df.loc[df['winner_seed'] == 'WC', 'winner_seed'] = float('NaN')
```

```python
# Adjusting values LL, Q and WC to be loser_entry values instead of loser_seed
df.loc[df['loser_seed'] == 'LL', 'loser_entry'] = 'LL'
df.loc[df['loser_seed'] == 'LL', 'loser_seed'] = float('NaN')
df.loc[df['loser_seed'] == 'Q', 'loser_entry'] = 'Q'
df.loc[df['loser_seed'] == 'Q', 'loser_seed'] = float('NaN')
df.loc[df['loser_seed'] == 'WC', 'loser_entry'] = 'WC'
df.loc[df['loser_seed'] == 'WC', 'loser_seed'] = float('NaN')
```

```python
# Changing winner and loser seed column to numeric
df['winner_seed']= pd.to_numeric(df['winner_seed'], errors='coerce')
df['loser_seed']= pd.to_numeric(df['loser_seed'], errors='coerce')
```

```python
df['winner_entry'].value_counts()
```

```python
# Adjusting to all caps for Alt => ALT for winner and loser entry
df.loc[(df['loser_entry'] == 'Alt', 'loser_entry')] = 'ALT'
df.loc[(df['winner_entry'] == 'Alt', 'winner_entry')] = 'ALT'
```

```python
# Adjusting S to SE
df.loc[(df['loser_entry'] == 'S', 'loser_entry')] = 'SE'
```

```python
# Checking result
df['winner_entry'].value_counts()
```

```python
# Recheck for mixed types to see if issues have been resolved
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```
